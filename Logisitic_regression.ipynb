```python
#Importing lib
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 

# Reading data
df = pd.read_csv("C:/Users/Kotti/Downloads/data.csv")
df.info()
```

```python
df.diagnosis 
```

```python
# making label for diagnosis 

ln=LabelEncoder()

df.diagnosis=ln.fit_transform(df['diagnosis'])

df.diagnosis
```

```python
#Finding correlation
df.corr()
```

```python
# Assigning inputs and output
X_features={'radius_mean','perimeter_mean','concave points_worst','concave points_mean','perimeter_worst','area_worst'}
y=df['diagnosis']
```

```python
X = df[X_features]
X
```

```python
df.isna().sum()
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
```

```python

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=51)
```

```python
 
classifier= LogisticRegression(random_state=10)  
```

```python
classifier.fit(x_train, y_train)  
```

```python
#Predicting the test set result  
y_pred= classifier.predict(x_test)  
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
#Creating the Confusion matrix  
from sklearn.metrics import confusion_matrix  
confusion_matrix(y_test,y_pred)  
```

```python
#Hypertuning

from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
```

```python
parameters={'penalty':('l1', 'l2','elasticnet'),'C':[1,10,20,30]}
```

```python
classifier=LogisticRegression()
```

```python
clf=GridSearchCV(classifier,param_grid=parameters,cv=5)
clf.fit(x_train,y_train)
```

```python
clf.best_score_
```

```python
clf.best_params_
```

```python
# Passing best parameters into model
classifier=LogisticRegression(C=30,penalty='l2')
classifier.fit(x_train,y_train)
```

```python
#Predicting the test set result  
y_pred= classifier.predict(x_test)  
y_pred
```

```python
#Creating the Confusion matrix  
from sklearn.metrics import confusion_matrix  
confusion_matrix(y_test,y_pred)  
```

```python
# After data hypertuning the score of test data 
accuracy_score(y_test, y_pred)
```
